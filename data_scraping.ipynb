```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
def get_next_mehr_news(soup):
    footer = soup.findAll('a', attrs = {"class": "page-link"})
    next_link = None
    for link in footer:
        if link.text == 'بعدی':
            next_link = link['href']
    return next_link
```

```python
MAX_news = 3000
sport_news = []
```

```python
news=[]
BASE_URL ='https://www.mehrnews.com/'
URL = BASE_URL + 'archive?pi=1&tp=9&ms=0&dy=9&mn=9&yr=1402'
while True:
    req = requests.get(URL)
    soup = BeautifulSoup(req.content, 'html.parser')
    table = soup.findAll('div', attrs = {'class':"desc"})
    for row in table:
        if row.h3 != None and row.h3.a.text != '\n':
            news.append(row.h3.a.text)
    next_link = get_next_mehr_news(soup)
    if (next_link is None) or (len(news) > MAX_news):
        break
    URL = BASE_URL + next_link
sport_news += news
```
